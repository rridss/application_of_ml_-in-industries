```python
import pandas as pd
data = pd.read_csv("/content/indian_liver_patient.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis = 1)
data.drop_duplicates(inplace=True)
data
```

```python
print(data['Gender'].unique())

```

```python
# Replace non-numeric values in the 'gender' column with numeric values
data.replace({'Gender': {'Male': 0, 'Female': 1}}, inplace=True)

# Convert the 'gender' column to numeric
data['Gender'] = pd.to_numeric(data['Gender'], errors='coerce')

# Calculate the correlation with the 'is_patient' column
corr_values = data.corr()['Total_Bilirubin'].abs().sort_values(ascending=False)
print(corr_values)
```

```python
X = data.drop(columns = ['is_patient']).values
Y = data['']
```

```python
from sklearn.model_selection import train_test_split
XTrain, XTest, YTrain, YTest = train_test_split(X,Y, test_size= 0.20, random_state= 2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors= 4)
model.fit(XTrain, YTrain)
```
